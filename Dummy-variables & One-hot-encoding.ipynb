####Dummy variables part

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Locations_with_prices.csv')
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

###merge the dummy data to our original dataFrame or data table

```python
merged= pd.concat([df,dummies],axis='columns')
merged
```

##We will drop one of the dummy column for best result and also our town columns because we will give the numeric data to our model

```python
final = merged.drop(['town', 'lorlai'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

##Drop the price column

```python
X = final.drop('price' , axis= 'columns')
X
```

```python
y = final.price
y
```

```python
model.fit(X,y)
```

##Here i want to make 1,0 for predicting the sherani price. same as if you want to print the price of zhob then 0,1 after area in predict method.

```python
model.predict([[2000,1,0]])   #for sherani
```

```python
model.predict([[2000,0,1]])     # for zhob
```

```python
model.predict([[2000,0,0]])    #waist winser where both are zer0
```

##Accuracy of the predicated values

```python
model.score(X,y)
```

###One hot encoding part

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Locations_with_prices.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

###To convert the town names into numeric data

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town','area']].values
```

```python
X
```

```python
y = dfle.price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')   ##First column into one hot encodeing
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
```

```python
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,3400]]) # 3400 sqr ft home in zhob
```

```python
model.score(X,y)
```
