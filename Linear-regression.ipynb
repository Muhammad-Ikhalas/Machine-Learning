```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv('Home.csv')
```

```python
print(df)
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price, color='red',marker='+')
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```python
import numpy as np

# Assuming reg is your trained LinearRegression model
input_value = 1800
input_array = np.array([[input_value]])

# Now you can use predict with the reshaped input
prediction = model.predict(input_array)
```

```python
prediction
```

###Now to save the model to a file using pickle

```python
import pickle
with open('modle_pickle', 'wb') as f:
  pickle.dump(model,f)
```

###Loading and make prediction to saved model

```python
with open('/content/modle_pickle','rb') as f:
  mp=pickle.load(f)
```

```python
import numpy as np

# Assuming 'loaded_model' is the model you loaded
input_value = 1800
input_array = np.array([[input_value]])

# Use the loaded model to make predictions
prediction = mp.predict(input_array)

prediction
```
