```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
# To import iris dataset
iris = datasets.load_iris()
X = iris.data[:,:]   
y = iris.target
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.5, random_state=25)
knn = KNeighborsClassifier(n_neighbors=1)
dt = DecisionTreeClassifier(random_state=0)
gnb = GaussianNB()
knn.fit(x_train, y_train)
dt.fit(x_train, y_train)
gnb.fit(x_train, y_train)
y_predsknn = knn.predict(x_test)
y_predsdt = dt.predict(x_test)
y_predsgnb = gnb.predict(x_test)
# Knn, Dt and Ngb Models
print('Accuracy of Knn on iris is  {}'.format(accuracy_score(y_test, y_predsknn)))
print('Accuracy of Decision tree on iris is {}'.format(accuracy_score(y_test, y_predsdt)))
print('Accuracy of Naive Bayes on iris is {}'.format(accuracy_score(y_test, y_predsgnb)))
# Ensumble with hard voting
eknn_dt_gnb = VotingClassifier(estimators=[('knn', knn), ('dt', dt), ('gnb', gnb)], voting='hard')
eknn_dt_gnb = eknn.fit(x_train, y_train)
eknn_dt_gnb_preds = eknn.predict(x_test)
print('Accuracy of Ensemble classifiers with hard voting is {}'.format(accuracy_score(y_test, eknn_preds)))
# Ensumble with soft voting
eknn_dt_gnb = VotingClassifier(estimators=[('knn', knn), ('dt', dt), ('gnb', gnb)], voting='soft')
eknn_dt_gnb = eknn.fit(x_train, y_train)
eknn_dt_gnb_preds = eknn.predict(x_test)
print('Accuracy of Ensemble classifiers with soft voting is {}'.format(accuracy_score(y_test, eknn_preds)))
```
