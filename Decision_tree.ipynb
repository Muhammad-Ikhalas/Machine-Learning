```python
import pandas as pd
df = pd.read_csv('/content/salary_detail.csv')
df.head()
```

###Drop the target column

```python
inputs = df.drop('salary_more_then_100k' ,axis= 'columns')
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

##Label_encoder for the company job and degree because the model is not understand non-numeric data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree '])    # the degree name has an extra space at the end if you remove it than there will be an error
inputs.head()
```

###Now drops the non njmeric data columns

```python

inputs_n = inputs.drop(['company','job','degree '],axis='columns')
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n, target)
```

```python
model.predict([[2,2,1]])   # google	sales executive	master
```
