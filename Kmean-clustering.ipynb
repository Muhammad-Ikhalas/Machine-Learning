```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/content/Salary_list.csv')
df.head()
```

```python
plt.xlabel('Age')
plt.ylabel('income')
plt.scatter(df['Age'],df['Income'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicated = km.fit_predict(df[['Age','Income']])
y_predicated
```

```python
df['cluster'] = y_predicated
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.Age,df1['Income'],color='green')
plt.scatter(df2.Age,df2['Income'],color='red')
plt.scatter(df3.Age,df3['Income'],color='blue')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
```

###Our one group is classified preety good but other 2 is good classified therefore we should preprocessed our data

```python
scaler = MinMaxScaler()       ##the values will become between 0 and 1
scaler.fit(df[['Income']])
df['Income'] = scaler.transform(df[['Income']])
df
```

```python
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
df
```

```python
km = KMeans(n_clusters=3)
y_predected  = km.fit_predict(df[['Age','Income']])
y_predicated
```

```python
df['Cluster'] = y_predicated
df.drop('cluster' , axis='columns', inplace=True)
df
```

###Centeriods

```python
km.cluster_centers_
```

```python
df1 = df[df.Cluster==0]
df2 = df[df.Cluster==1]
df3 = df[df.Cluster==2]
plt.scatter(df1.Age,df1['Income'],color="green")
plt.scatter(df2.Age,df2['Income'],color="red")
plt.scatter(df3.Age,df3['Income'],color="blue")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='+',label='centroid')
plt.legend()
```

###Elbow technique

```python
k_range = range(1,10)
sse = []
for k in k_range:
  km = KMeans(n_clusters=k)
  km.fit(df[['Age','Income']])
  sse.append(km.inertia_)
```

```python
sse   ##SSE ERROR
```

```python
plt.xlabel('K')
plt.ylabel('Sum of the squared error')
plt.plot(k_range,sse)
```
