```python
from sklearn.datasets import load_iris
iris = load_iris()
iris.target
X= iris.data
y= iris.target
```

**Split the data to testing and traning**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
len(X_train) #   =120
len(X_test)  #   =30
```

**KNN CLASSIFICATION STARTS FROM HERE**


```python
# KNN starts from here
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
# print(knn)
# The ‘fit’ method trains the algorithm on the training data
knn.fit(X_train,y_train)
```

**To print the accuracy**

```python
print(formate(knn.score(X_test, y_test))
```
